```python
import pandas as pd
```

```python
df1 = pd.read_csv("defect_by_operno.csv")
df1
```

```python
n = df1['WAFER_ID'].nunique()
print(n)

bad_wafer = df1['WAFER_ID'].unique()[:65200]
print(len(bad_wafer))

good_wafer = df1['WAFER_ID'].unique()[65200:]
print(len(good_wafer))
```

```python
df1['label'] = df1['WAFER_ID'].apply(lambda x: 1 if x in bad_wafer else 0)
```

```python
df1.to_csv("defect_by_operno.csv.go", index=0)
```
