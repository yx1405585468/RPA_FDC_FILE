```python
grpby_list_ = ['OPE_NO', 'CHAMBER_NAME', 'PRODUCT_ID', 'EQP_NAME']
```

```python
df = df_pandas.copy()
df['PARAM'] = df['PARAMETRIC_NAME'].apply(lambda x: x.split('#')[0])
df['STEP'] = df['PARAMETRIC_NAME'].apply(lambda x: x.split('#')[1])
df['STATS'] = df['PARAMETRIC_NAME'].apply(lambda x: x.split('#')[2])
```

```python
df 
```

```python
grpby_list_extend = grpby_list_ + ['PARAM', 'STEP']
feature_stats = df.groupby(grpby_list_extend)['STATS'].unique().reset_index()
feature_stats
```

```python
def count_good_bad(group):
    good_num = group[group['label'] == 0]['WAFER_ID'].nunique()
    bad_num = group[group['label'] == 1]['WAFER_ID'].nunique()
    return pd.Series([good_num, bad_num], index=['GOOD_NUM', 'BAD_NUM'])

count_good_bad_df = df.groupby(grpby_list_extend).apply(count_good_bad).reset_index()
count_good_bad_df
```
