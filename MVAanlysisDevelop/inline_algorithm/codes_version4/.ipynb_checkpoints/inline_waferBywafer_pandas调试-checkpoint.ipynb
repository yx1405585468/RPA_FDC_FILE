```python
def select_columns(df, columns_list):
    return df[columns_list]


def get_label(x, good_wafer, bad_wafer):
    if x in good_wafer:
        return 0
    elif x in bad_wafer:
        return 1
    else:
        return 'Nolabel'

    
def run_get_label(df, good_wafer, bad_wafer):
    df['label'] = df['WAFER_ID'].apply(lambda x: get_label(x, good_wafer, bad_wafer))
    df = df[df['label'] != 'Nolabel']
    return df 
```

```python
def exclude_some_data(df, certain_column: str, key_words:list[str]):
    key_words_str = '|'.join(key_words)
    df = df[~df[certain_column].str.contains(key_words_str)]
    return df
```

```python
def pre_process(df, convert_to_numeric_list):
    
    # 数据类型转换
    df[convert_to_numeric_list] = df[convert_to_numeric_list].apply(pd.to_numeric)
    
    # 七个统计值都为NA时，删除这些数据
    if 'SITE_COUNT' in convert_to_numeric_list:
        convert_to_numeric_list.remove('SITE_COUNT')
    df = df.dropna(subset=convert_to_numeric_list, axis=0, how='all')

    # 缺失值填充
#     new_df[agg_list] = new_df[agg_list].apply(lambda x: x.fillna(x.mean()))

    return df
```

```python
def process_missing_values(df, columns_to_process, threshold=0.6):
    df_processed = df.copy()
    for column in columns_to_process:
        # 计算当前列的缺失值比例
        missing_percentage = df[column].isnull().mean()
        # 如果缺失值比例超过阈值，删除该列；否则用均值填充
        if missing_percentage > threshold:
            df_processed = df_processed.drop(columns=[column])
        else:
            df_processed[column] = df_processed[column].fillna(df_processed[column].mean())
    return df_processed
```

```python
res_top_select_all = pd.DataFrame()

for i in range(len(operno_list)): 
    oper = operno_list[i]
    print(f"*************当前站点是{oper}*************")
    df_pandas_select_specific_operno = df_pandas_select.query(f"OPE_NO == '{oper}'")
    
    columns_to_process = ['AVERAGE', 'MAX_VAL', 'MEDIAN', 'MIN_VAL', 'STD_DEV', 'PERCENTILE_25', 'PERCENTILE_75']
    df_pandas_select_specific_operno = process_missing_values(df=df_pandas_select_specific_operno, 
                                                              columns_to_process=columns_to_process, 
                                                              threshold=0.6)

    values_list = df_pandas_select_specific_operno.columns.difference(['WAFER_ID', 'OPE_NO', 'INLINE_PARAMETER_ID', 'SITE_COUNT', 'label'])
    pivot_result = df_pandas_select_specific_operno.pivot_table(index=['WAFER_ID', 'label'], 
                                                                columns=['OPE_NO', 'INLINE_PARAMETER_ID'],
                                                                values=values_list)
    pivot_result.columns = pivot_result.columns.map('#'.join)
    pivot_result = process_missing_values(df=pivot_result, columns_to_process=pivot_result.columns, threshold=0.6)
    pivot_result = pivot_result.reset_index(drop=False)

    # 定义自变量
    df_pivot_all = pivot_result.copy()
    
    x_train = df_pivot_all[df_pivot_all.columns.difference(['WAFER_ID', 'label']).tolist()]
    print("x_train维度是", x_train.shape)
    
    n_components = min(min(x_train.shape)-1, 5)
    print("降维个数是", n_components)
    
    model = pca(n_components=n_components, verbose=None)
    results = model.fit_transform(x_train)
    res_top = results['topfeat']
    res_top_select = res_top[res_top['type'] == 'best'][['feature', 'loading']]
    res_top_select = res_top_select.drop_duplicates()
    res_top_select['importance'] = abs(res_top_select['loading'])
    res_top_select = res_top_select.rename(columns={'feature': 'features'})
    res_top_select = res_top_select.drop("loading", axis=1)
    res_top_select_all = pd.concat([res_top_select_all, res_top_select], axis=0)
```

```python
def split_features(df, index) -> str:
    """
    param df: RandomForest建模后的feature_importance_table
    param index: 顺序值
    return: 字段属性值
    """
    return df['features'].apply(lambda x: x.split('#')[index])
```

```python
def get_split_feature_importance_table(df):
    df['STATISTIC_RESULT'] = split_features(df, 0)
    df['OPER_NO'] = split_features(df, 1)
    df['INLINE_PARAMETER_ID'] = split_features(df, 2)
    
    df = df.drop(['features', 'STATISTIC_RESULT'], axis=1).reset_index(drop=True)
    return df
```

```python
split_table = get_split_feature_importance_table(res_top_select_all) 
print(split_table.shape)
split_table_grpby = split_table.groupby(['OPER_NO', 'INLINE_PARAMETER_ID'])['importance'].sum().reset_index(drop=False)
split_table_grpby.shape
```
